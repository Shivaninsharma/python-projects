```python
from tkinter import *
import math
root = Tk()
root.title("Calculator works on Two numbers")

#values = IntVar
user_text = Entry(root,width = 35,borderwidth = 5)
user_text.grid(row = 0, column = 0, columnspan = 5, padx = 10 , pady = 10)

def button_click(n):
    user_text.insert(0,n)

def add():
    no = user_text.get()
    global f_ad_no
    f_ad_no = int(no)
    global maths 
    maths  = "addition"
    user_text.delete(0,END)

def sub():
    no = user_text.get()
    global f_sb_no
    f_sb_no = int(no)
    global maths
    maths  = "subtraction"
    user_text.delete(0,END)

def mul():
    no = user_text.get()
    global f_ml_no
    f_ml_no = int(no)
    global maths
    maths = "multiple"
    user_text.delete(0,END)

def div():
    no = user_text.get()
    global f_dv_no
    f_dv_no = int(no)
    global maths 
    maths = "division"
    user_text.delete(0,END)

def button_sin():
    no = user_text.get()
    global sin_no
    sin_no = int(no)
    global maths
    maths = "sin"
    user_text.delete(0,END)

def button_cos():
    no = user_text.get()
    global cos_no
    cos_no = int(no)
    global maths
    maths = "cos"
    user_text.delete(0,END)

def button_square():
    no = user_text.get()
    global sq_no
    sq_no = int(no)
    global maths
    maths = "square"
    user_text.delete(0,END)

def button_root():
    no = user_text.get()
    global root_no
    root_no = no
    global maths
    maths = "root"
    user_text.delete(0,END)

def cube():
    no = user_text.get()
    global cu_no
    cu_no = int(no)
    global maths
    maths = "cube"
    user_text.delete(0,END)
    
def button_clear():
    user_text.delete(0,END)

def button_equal():
    sec_no = user_text.get()
    user_text.delete(0,END)
    if maths  == "addition":
        user_text.insert(0,f_ad_no+int(sec_no))
    if maths == "subtraction":
        user_text.insert(0,f_sb_no+int(sec_no))
    if maths == "multiple":
        user_text.insert(0,f_ml_no*int(sec_no))
    if maths == "division":
        user_text.insert(0,f_dv_no/int(sec_no))
    if maths == "sin":
        user_text.insert(0,math.sin(sin_no))
    if maths == "cos":
        user_text.insert(0,math.cos(cos_no))
    if maths == "square":
        user_text.insert(0,(sq_no)**2)
    if maths == "root":
        user_text.insert(0,math.sqrt(root_no))
    if maths == "cube":
        user_text.insert(0,cu_no**3)
    


Button(root,text = "1",padx = 20,pady = 10, command = lambda : button_click(1)).grid(row = 1, column = 0)
Button(root,text = "2",padx = 20,pady = 10, command = lambda : button_click(2)).grid(row = 1, column = 1)
Button(root,text = "3",padx = 20,pady = 10, command = lambda : button_click(3)).grid(row = 1, column = 2)
Button(root,text = "4",padx = 20,pady = 10, command = lambda : button_click(4)).grid(row = 1, column = 3)
Button(root,text = "5",padx = 20,pady = 10, command = lambda : button_click(5)).grid(row = 1, column = 4)
Button(root,text = "6",padx = 20,pady = 10, command = lambda : button_click(6)).grid(row = 2, column = 0)
Button(root,text = "7",padx = 20,pady = 10, command = lambda : button_click(7)).grid(row = 2, column = 1)
Button(root,text = "8",padx = 20,pady = 10, command = lambda : button_click(8)).grid(row = 2, column = 2)
Button(root,text = "9",padx = 20,pady = 10, command = lambda : button_click(9)).grid(row = 2, column = 3)
Button(root,text = "0",padx = 20,pady = 10, command = lambda : button_click(0)).grid(row = 2, column = 4)
Button(root,text = "+",padx = 20,pady = 10, command = add).grid(row = 3, column = 0)
Button(root,text = "-",padx = 20,pady = 10, command = sub).grid(row = 3, column = 1)
Button(root,text = "*",padx = 20,pady = 10, command = mul).grid(row = 3, column = 2)
Button(root,text = "/",padx = 20,pady = 10, command = div).grid(row = 3, column = 3)
Button(root,text = "=",padx = 20,pady = 10, command = button_equal).grid(row = 3, column = 4)
Button(root,text = "sin",padx =20, pady = 10, command = button_sin).grid(row =4, column = 0)
Button(root,text = "cos",padx =20, pady = 10, command = button_cos).grid(row =4, column = 1)
Button(root,text = "x^2",padx =20, pady = 10, command = button_square).grid(row =4, column = 2)
Button(root,text = "root(x)",padx =20, pady = 10, command = button_root).grid(row =4, column = 3)
Button(root,text = "x^3",padx =20, pady = 10, command = cube).grid(row =4, column = 4)
Button(root,text = "clear",padx = 70,pady = 10,command = button_clear).grid(row = 5, column = 0,columnspan = 5)




root.mainloop()
```
